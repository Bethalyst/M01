```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array ([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])
plt.plot(x, y, 'o:y')
font1 = {'family':'serif','color':'m','size':20}
plt.title("Simple Line Plot", fontdict = font1 )
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values, color = '#9932CC')
font1 = {'family':'Lucida Console','color':'k','size':20}
plt.title("Category Bar Chart", fontdict = font1)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [20, 30, 25, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Simple Pie Chart")
plt.axis('equal') 
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
data = { 'Product': ['A', 'B', 'A', 'B', 'C', 'A', 'C'], 'Region': ['North', 'South', 'South', 'North', 'North', 'South', 'South'], 'Sales': [100, 150, 200, 130, 90, 170, 50] }
df = pd.DataFrame(data)
df
```

```python
pivot_table = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot_table)
```

*This pivot table aggrigate and summarize a tabular data in a simple easy to read information*
